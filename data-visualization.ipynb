```python
# import libraries
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# path in data
data = pd.read_csv(r'C:\Users\awd51\Python Projects\portfolio/data-visualization\countries.csv')

data
```

```python
# create data frames for the United States and China
us = data[data.country == 'United States']
china = data[data.country == 'China']

print(type(us.population))
```

```python
# plot data
plt.plot(us.year, us.population / 10 ** 6)
plt.plot(china.year, china.population / 10 ** 6)
plt.legend(['United States', 'China'])
plt.xlabel('Year')
plt.ylabel('Population (Millions)')

plt.show()
```

```python
# plot data with points
plt.plot(us.year, us.population / us.population.iloc[0] * 100, 'o-', color = 'blue')
plt.plot(china.year, china.population / china.population.iloc[0] * 100, 'o-', color = 'red')
plt.legend(['United States', 'China'])
plt.xlabel('Year')
plt.ylabel('Population Growth (%)')

plt.show()
```
